```python
import pandas as pd
df = pd.read_csv('../datasets/titanic/titanic_cleaned.csv')
```

```python
# mean
mean_age = df['Age'].mean()
print("Mean age: ", mean_age)

mean_fare = df['Fare'].mean()
print("Mean fare: ", mean_fare)
```

```python
# variance
variance_age = df['Age'].var()
print("variance age: ", variance_age)

variance_fare = df['Fare'].var()
print("variance fare: ", variance_fare)
```

```python
# Co-variance
cov_matrix = df[['Age', 'Fare', 'Pclass']].cov()
cov_matrix
```

```python
# Co-relation
corr_matrix = df[['Age', 'Fare', 'Pclass']].corr()
corr_matrix
```

```python
# Grouped Statistics
df_new = df[['Survived', 'Age', 'Fare', 'Pclass']]
df_new.groupby('Pclass').mean()
```
