```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../datasets/titanic/titanic_cleaned.csv')

# ============ MEAN ============
print("="*50)
print("MEAN (Average Value)")
print("="*50)
mean_age = df['Age'].mean()

# ============ VARIANCE ============
print("\n" + "="*50)
print("VARIANCE (Spread of Data)")
print("="*50)
var_age = df['Age'].var()
var_fare = df['Fare'].var()
print(f"Variance Age: {var_age:.2f} (std dev: {df['Age'].std():.2f})")
print(f"Variance Fare: {var_fare:.2f} (std dev: {df['Fare'].std():.2f})")

# ============ COVARIANCE ============
print("\n" + "="*50)
print("COVARIANCE (Relationship Between Variables)")
print("="*50)
# Calculate covariance matrix for numeric columns
cov_matrix = df[['Age', 'Fare', 'Pclass']].cov()
print("\nCovariance Matrix:")
print(cov_matrix)

# Specific covariance
cov_age_fare = df['Age'].cov(df['Fare'])
print(f"\nCovariance(Age, Fare): {cov_age_fare:.2f}")

# ============ CORRELATION (Normalized Covariance) ============
print("\n" + "="*50)
print("CORRELATION (Normalized, -1 to 1)")
print("="*50)
corr_matrix = df[['Age', 'Fare', 'Pclass']].corr()
print(corr_matrix)

# Visualize
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(df['Age'].dropna(), bins=20, alpha=0.7, label=f'Mean={mean_age:.1f}')
plt.axvline(mean_age, color='red', linestyle='--', label='Mean')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution (showing mean)')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title(f'Age vs Fare (Correlation: {cov_age_fare:.2f})')
plt.tight_layout()
plt.show()
```
