```python
import pandas as pd
from scipy.spatial.distance import euclidean, cityblock, cosine
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv('../datasets/titanic/titanic_cleaned.csv')

df = df[['Survived', 'Age', 'Fare', 'Pclass', 'SibSp', 'Parch']]
df.groupby('Survived').mean()
```

```python
# Distance between two passengers
p1 = df.iloc[0][['Age', 'Fare', 'SibSp']].values
p2 = df.iloc[50][['Age', 'Fare', 'SibSp']].values
df.iloc[[0,50]][['Age', 'Fare', 'SibSp']]
```

```python
print('Euclidean Distance:', euclidean(p1, p2))
print('Manhattan Distance:', cityblock(p1, p2))
print('Cosine Distance:', cosine(p1, p2))
```

```python
# Scaling and Normalization
scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[['Age', 'Fare', 'SibSp']] = scaler.fit_transform(df[['Age', 'Fare', 'SibSp']])

p1_scaled = df_scaled.iloc[0].values
p2_scaled = df_scaled.iloc[50].values
df_scaled.iloc[[0,50]]
```

```python
print('Euclidean Distance (Scaled):', euclidean(p1_scaled, p2_scaled))
print('Manhattan Distance (Scaled):', cityblock(p1_scaled, p2_scaled))
print('Cosine Distance (Scaled):', cosine(p1_scaled, p2_scaled))
```
