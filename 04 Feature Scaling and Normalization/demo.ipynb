```python
import pandas as pd
from scipy.spatial.distance import euclidean, cityblock, cosine
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv('../datasets/titanic/titanic_cleaned.csv')
df[['Name', 'Age']].head()
```

```python
df = df[['Survived', 'Age', 'Fare', 'Pclass']]
df.groupby('Survived').mean()
```

```python
p1 = df.iloc[0].values
p2 = df.iloc[50].values

df.iloc[[0, 50]]
```

```python
print('Euclidean distance:', euclidean(p1, p2))
print('Manhattan distance:', cityblock(p1, p2))
print('Cosine distance:', cosine(p1, p2))
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['Age', 'Fare', 'Pclass']])
df_scaled = pd.DataFrame(df_scaled, columns=['Age', 'Fare', 'Pclass'])
```
