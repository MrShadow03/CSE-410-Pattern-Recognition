```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/titanic/titanic_cleaned.csv')

df.describe()
```

```python
df = df[['Survived', 'Age', 'Fare', 'Pclass', 'SibSp', 'Parch']]
```

```python
p1 = df.iloc[0]
p2 = df.iloc[50]

df.iloc[[0, 50]]
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = StandardScaler()
scaled_df = scaler.fit_transform(df[['Age', 'Fare', 'Pclass']])
scaled_df = pd.DataFrame(scaled_df, columns=['Age', 'Fare', 'Pclass'])

scaled_df.head()
```

```python
from scipy.spatial.distance import euclidean, cityblock, cosine

p1_scaled = scaled_df.iloc[0].values
p2_scaled = scaled_df.iloc[50].values

print('Euclidean distance:', euclidean(p1_scaled, p2_scaled))
print('Manhattan distance:', cityblock(p1_scaled, p2_scaled))
print('Cosine distance:', cosine(p1_scaled, p2_scaled))
```
