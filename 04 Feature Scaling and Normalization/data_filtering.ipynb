```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
#Loading dataset from csv file
file_path = '../datasets/titanic/train.csv'
df = pd.read_csv(file_path)
```

```python
# First few and last few rows of the dataset
print(df.head())
print(df.tail())
```

```python
# Viewing summary and statistics
print(df.info())
print(df.describe())
```

```python
# Viewing specific columns
df[['Age', 'Fare', 'Body']].head()
```

```python
# Filtering rows based on conditions
# Example: Passengers who survived and paid more than 50 in fare
filtered_df = df[(df['Survived'] == 1) & (df['Fare'] > 50)]
print(filtered_df[['Name', 'Age', 'Survived', 'Fare']].head())
```

```python
# Example 02: Body is not null and age is greater than 30
df[df['Body'].notnull()][['Name', 'Age', 'Body']].head()
```

```python
# Handle missing values
numeric_features = ['Age', 'SibSp', 'Parch', 'Fare']
print(df[df['Age'].isnull()][['Name', 'Age']].head())
df_numeric = df[numeric_features].copy()
df_numeric = df_numeric.fillna(df_numeric.mean())

df_numeric.head()
```
