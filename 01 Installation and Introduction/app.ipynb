```python
# List in python
ages = [25, 30, 35, 40, 45, 55.8, "Hello"]
print(ages[1]) # Accessing an element from the list
print(ages[2:]) # Slicing the list from index 2 to the end
ages.append(101) # Appending an element to the end of the list
ages.remove(35) # Removing an element from the list
print(ages)

# multidimensional list
matrix = [
    # temp, humidity, pressure
    [1, 2, 3], # day 1
    [4, 5, 6], # day 2
    [7, 8, 9] # day 3
]

import numpy as np
matrix = np.array(matrix)

print(matrix[1:,1])
```
