```python
from scipy.spatial.distance import euclidean, cityblock, cosine
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../datasets/titanic/titanic_cleaned.csv')
df_scaled = pd.read_csv('../datasets/titanic/titanic_scaled.csv')

# Euclidean distance between two passengers
passenger1 = df.iloc[0][['Age', 'Fare']].values
passenger2 = df.iloc[1][['Age', 'Fare']].values

passenger1_scaled = df_scaled.iloc[0][['Age', 'Fare']].values
passenger2_scaled = df_scaled.iloc[1][['Age', 'Fare']].values

euclidean_distance = euclidean(passenger1, passenger2)
print(f'Euclidean Distance: {euclidean_distance}')

euclidean_distance_scaled = euclidean(passenger1_scaled, passenger2_scaled)
print(f'Euclidean Distance (Scaled): {euclidean_distance_scaled}')
```
