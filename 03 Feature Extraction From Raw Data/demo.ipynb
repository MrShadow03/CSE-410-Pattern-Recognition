```python
import numpy as np

# 2D list and numpy array
data = [
    [23, 26, 30], #temperature
    [0.5, 0.6, 0.8], #humidity,
    [0, 0, 1] #rain
]
data = np.array(data)

print(data[:, 0])
print(data.shape)
```

```python
# Extract data from numeric samples
raw_data = [22.2, 0.4, 22.8, 24.2, 0.5, 23.5, 25.1, 0.6, 24.8]

# Mean
mean = np.mean(raw_data)
print("Mean:", mean)

# Standard Deviation
std = np.std(raw_data)
print("Standard Deviation:", std)

# Min
min_value = np.min(raw_data)
print("Min:", min_value)

# Max
max_value = np.max(raw_data)
print("Max:", max_value)

# Range
data_range = max_value - min_value
print("Range:", data_range)
```

```python
# Extract data from an image
import cv2

img = cv2.imread('../cliff.jpg')
# print(img)

# Hight and width
h, w, c = img.shape

red_channel = np.mean(img[:, :, 2])
print("Red channel mean:", red_channel)

green_channel = np.mean(img[:, :, 1])
print("Green channel mean:", green_channel)

blue_channel = np.mean(img[:, :, 0])
print("Blue channel mean:", blue_channel)



bw_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

brightness = np.mean(bw_img)
print("Brightness:", brightness)

contrast = np.std(bw_img)
print("Contrast:", contrast)
```

```python
# Extract data from a video
cap = cv2.VideoCapture('../sample.mp4')

frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
print("Frame count:", frame_count)

fps = cap.get(cv2.CAP_PROP_FPS)
print("Frames per second:", fps)

duration = frame_count / fps
print("Duration (seconds):", duration)
```
